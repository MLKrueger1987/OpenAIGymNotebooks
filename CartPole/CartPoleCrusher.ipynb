# Agent Testing Function
### Runs the Agent 100 times and returns the mean score of the 100 trials

```python
def test(learner):
    rewards = []
    for i in range(100):
        reward = 0
        s=torch.tensor(env.reset()).float()
        done = False
        while not done:
            a = torch.tensor(learner.predict(s))
            sprime, r, done, _= env.step(a.item())
            sprime = torch.tensor(sprime).float()
            reward+=r
            r = torch.tensor(r)
            done = torch.tensor(done)
            s=sprime
        rewards.append(reward)
    return np.mean(rewards)
```

# Agent Training

```python
alpha =.001
gamma = .99
tau = .8
max_decay = 50
init_eps =1
final_eps =.005

learner = DQN.DQNLearner(alpha,gamma,tau,num_states,num_actions,
                         memory_size=5000,
                         batch=64,max_iter=61000,
                         max_decay=max_decay,init_eps=init_eps,
                         final_eps=final_eps,device='cuda')
steps = 200 # Assume 500 steps per episode, then multiply by 
# number of desired episodes
# learner.set_memory(70000,32)
actions = []
scores = []
avg_scores = []
print('Starting:')
e = 0
solved = False
loss = []
for i in range(500):
    s = torch.tensor(env.reset()).float()
    done = False
    render = True
    reward = 0
    

    while not done:
#         env.render()
        a = learner.query(s)

        sprime, r, done, _= env.step(a.item())
        sprime = torch.tensor(sprime).float()
        reward+=r
        r = torch.tensor(r)
        done = torch.tensor(done)
#         if done and r > 0:
#             r *= 10
        learner.save_exp(s,a,r,sprime,done)
        if learner.memory.ready():
            learner.learn(s,a,r,sprime)
            loss.append(learner.lossval)
            
        s = sprime

    
#     if learner.memory.ready():
    learner.next_epoch()
    # end of episode
    scores.append(reward)
    if e % 25 == 0:
        solved = plot_scores(scores,loss,'tiny','tiny')
    render=False
    e+= 1
    
    if solved:
        break


solved = plot_scores(scores,loss,'tiny','tiny')
```
